```python
import pandas as pd
import numpy as np
import random

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
num_records = 500  # Number of accident records

# Date and Time (Randomly sampled from a specific range)
date_range = pd.date_range(start="2024-01-01", end="2024-12-31", freq="H")
dates = np.random.choice(date_range, num_records)

# Weather conditions
weather_conditions = ['Clear', 'Rain', 'Fog', 'Snow', 'Storm', 'Cloudy']
weather = np.random.choice(weather_conditions, num_records)

# Road conditions
road_conditions = ['Dry', 'Wet', 'Icy', 'Snowy', 'Under Construction']
road_condition = np.random.choice(road_conditions, num_records)

# Time of day
# Convert numpy.datetime64 to datetime objects before using strftime
time_of_day = [pd.to_datetime(date).strftime("%H:%M:%S") for date in dates]

# Accident severity
severity = np.random.choice([1, 2, 3, 4], num_records)

# Generate random locations (latitude and longitude)
latitudes = np.round(np.random.uniform(37.0, 39.0, num_records), 4)  # Latitude in some region
longitudes = np.round(np.random.uniform(-122.0, -120.0, num_records), 4)  # Longitude in some region

# Create DataFrame
accident_data = pd.DataFrame({
    'Date': [pd.to_datetime(date).strftime("%Y-%m-%d") for date in dates], # Convert to datetime objects
    'Time': time_of_day,
    'Weather': weather,
    'Road_Condition': road_condition,
    'Severity': severity,
    'Latitude': latitudes,
    'Longitude': longitudes
})

# Display first few rows
accident_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Count plot of accidents by weather condition
plt.figure(figsize=(7, ))
sns.countplot(data=accident_data, x='Weather', palette='magma')
plt.title('Accidents by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.show()
```
