```python
#Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('train.csv')
print("First few rows of the dataset:")
print(df.head())
print("\nMissing values in each column:")
print(df.isnull().sum())
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns=['Cabin'], inplace=True)
```

```python
#step 2: exploratory data analysis(EDA)


df = pd.read_csv('train.csv') # Load the DataFrame here
print("\nSummary statistics:")
print(df.describe())


#Bar plot for categorical features like 'pclass', 'sex', and 'Embarked'
plt.figure(figsize=(18,5))

#passenger class distribution
plt.subplot(1 ,3 ,1)
sns.countplot(x='Pclass',data=df)
plt.title('Passenger Class Distribution')

plt.subplot(1,3,2)
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')

#Embarktion port distribution
plt.subplot(1,3,3)
sns.countplot(x='Embarked', data=df)
plt.title('Embarkation Port Distribution')

plt.tight_layout()
plt.show()

#histogram for 'Age'
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

#box plot 'face' by 'pclass'
plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
